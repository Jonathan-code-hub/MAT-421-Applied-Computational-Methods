Euler's method:

```python
import numpy as np
import matplotlib.pyplot as plt

def euler_solve(f,a,b,n,y_0):

  """
  Parameters:
  f: function,
  a: float, intial x value
  b: float, final x value
  n: int, number of steps
  y_0: intial y value

  Output:
  y: float, final y value
  y_plot: array,
  """

  x = a
  dx = (b-a)/(n)
  y = y_0
  y_plot = np.array([])
  for i in range(n):
    y = y + f(y,x)*(dx)
    x += dx
    y_plot = np.append(y_plot,y)

  return y, y_plot

def dydx(y,x):
  return y**(2)*x + x

# Testing #
a = 0
b = 1
n = 100
y_0 = 1
x_plot = np.linspace(a,b,n)

y_f, y_plot = euler_solve(dydx,a,b,n,y_0)

plt.plot(x_plot, y_plot)
```

Improved Euler's Method:

```python
import numpy as np
import matplotlib.pyplot as plt

def improved_euler_solve(f, a, b, n, y_0):
    """
    Parameters:
    f: function, the first-order ordinary differential equation dy/dx = f(x, y)
    a: float, initial x value
    b: float, final x value
    n: int, number of steps
    y_0: initial y value

    Output:
    y: float, final y value
    y_plot: array, y values at each step
    """
    h = (b - a) / n  # Step size
    x = np.linspace(a, b, n+1)  # Array of x values
    y = np.zeros(n+1)  # Array to store y values
    y[0] = y_0  # Initial y value

    for i in range(n):
        k1 = f(x[i], y[i])
        k2 = f(x[i] + h, y[i] + h * k1)
        y[i+1] = y[i] + (h / 2) * (k1 + k2)

    return y[-1], y

# Testing #
def f(x, y):
    return y**(2)*x + x

a = 0
b = 1
n = 10
y_0 = 1
x_plot = np.linspace(a,b,n+1)

y_f, y_plot = improved_euler_solve(dydx,a,b,n,y_0)

plt.plot(x_plot, y_plot)
```

Runge Kutta 4 Method:

```python
import numpy as np
import matplotlib.pyplot as plt

def runge_kutta(f, a, b, n, y_0):
    h = (b - a) / n
    x = a
    y = y_0
    y_plot = [y_0]

    for i in range(n):
        k1 = h * f(x, y)
        k2 = h * f(x + 0.5 * h, y + 0.5 * k1)
        k3 = h * f(x + 0.5 * h, y + 0.5 * k2)
        k4 = h * f(x + h, y + k3)

        y += (k1 + 2 * k2 + 2 * k3 + k4) / 6
        x += h
        y_plot.append(y)

    return y, np.array(y_plot)

# Testing #
def f(x, y):
    return y**(2)*x + x

a = 0
b = 1
n = 10
y_0 = 1
x_plot = np.linspace(a,b,n+1)

y_f, y_plot = runge_kutta(dydx,a,b,n,y_0)

plt.plot(x_plot, y_plot)
```
